```python
#import the csv file
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
 #read the csv file
df = pd.read_csv(io.BytesIO(uploaded['data.csv']))
print(df)
```

```python
#Show the basic statistical description about the data
df.describe()
```

```python
#Check if the data has null values.
null=df.isnull().sum()
print(null)
```

```python
#replace null values with mean
df.fillna(df.mean(), inplace=True)
df
```

```python
df.isnull().sum()
```

```python
#Select at least two columns and aggregate the data using: min, max, count, mean
import pandas as pd
selected_columns =['Duration','Pulse']
aggregated=df[selected_columns].agg(['min','max','count','mean'])
print(aggregated)
```

```python
#Filter the dataframe to select the rows with calories values between 500 and 1000.
import pandas as pd
result=df[(df['Calories']>500) & (df['Calories']<1000)]
print(result)
```

```python
#Filter the dataframe to select the rows with calories values > 500 and pulse < 100
import pandas as pd
result=df[(df['Calories']>500) & (df['Pulse']<100)]
print(result)
```

```python
#Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”
import pandas as pd
df_new=df.drop('Maxpulse',axis=1)
print(df_new)
```

```python
#Delete the "Maxpulse" column from the main df dataframe
import pandas as pd
df.drop('Maxpulse',axis=1,inplace=True)
df
```
